```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
cars=pd.read_csv("cars.csv")
```

```python
cars
```

```python
x=[1,2,3,4,5]
y=[11,23,37,49,51]
plt.figure()
sns.lineplot(x,y)
plt.show()
```

```python
plt.figure()

sns.distplot(cars["width"])
plt.show()
```

# histogram

```python
plt.figure()

sns.distplot(cars["price"])
plt.show()
```

# Countplot

```python
plt.figure()

sns.countplot(data=cars,x="body-style")

plt.show()
```

```python
plt.figure()

sns.countplot(data=cars,y="body-style")

plt.show()
```

```python
plt.figure()

sns.countplot(data=cars,x="body-style",hue="fuel-type")

plt.show()
```

```python
plt.figure()

sns.countplot(data=cars,x="body-style",hue="engine-location")

plt.show()
```

# heatmap

```python
plt.figure()
sns.heatmap(cars.corr(),annot=True)

plt.show()
```

```python
#color palettes
plt.figure()
sns.heatmap(cars.corr(),annot=True,cmap="Greens")

plt.show()
```

```python
#scatter plot
```

```python
plt.figure()
sns.scatterplot(data=cars,x="engine-size",y="price")
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=cars,x="highway-mpg",y="price")
plt.show()
```

```python
#hue
plt.figure()
sns.scatterplot(data=cars,x="engine-size",y="price",hue="body-style")
plt.show()
```

# Box plot

# To plot outliers

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=cars,x="price")
plt.show()
```

```python
plt.figure(figsize=(5,8))

sns.boxplot(data=cars,y="price")
plt.show()
```

```python
plt.figure(figsize=(10,10))

sns.boxplot(data=cars,x="price",y="make")
plt.show()
```

```python
plt.figure(figsize=(10,10))

sns.boxplot(data=cars,x="price",y="body-style")
plt.show()
```

```python
##subplots
```

```python
fig,axes=plt.subplots(2,2,figsize=(8,8))

sns.countplot(data=cars,x="body-style",hue="fuel-type",ax=axes[0,0])
sns.scatterplot(data=cars,x="engine-size",y="price",ax=axes[0,1])
sns.boxplot(data=cars, x="price",y="body-style", ax=axes[1,0])
sns.heatmap(cars.corr(),annot=True, ax=axes[1,1])


plt.show()
```
