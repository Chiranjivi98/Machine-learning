```python
cars = pd.read_csv("cars.csv")
```

```python
plt.figure()
```

```python
plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.hist(cars["price"])
plt.show()
```

```python
price_mean = cars["price"].mean()
price_median = cars["price"].median()
```

```python
plt.figure()

plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.axvline(price_mean,color="green")
plt.axvline(price_median,color="red")

plt.hist(cars["price"])


plt.show()
```

```python
plt.figure()

plt.title("Width distribution")
plt.xlabel("width")
plt.ylabel("Frequency")

plt.hist(cars["width"])


plt.show()
```
